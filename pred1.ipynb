```python
import re
import pickle
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from tensorflow.keras.models import load_model

model = load_model("best_imdb_bilstm.keras")
```

```python
tokenizer = pickle.load(open("tokenizer.pickle", "rb"))
```

```python
MAXLEN = 200
```

```python
import nltk
from nltk.tokenize import word_tokenize


def preprocess(text):
    text = text.lower()
    text = re.sub(r'<[^<]+?>', '', text)                 # remove HTML
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)  # remove urls
    text = re.sub(r'[^\w\s]', '', text)                  # remove punctuation
    words = word_tokenize(text)
    return " ".join(words)
```

```python
samples = [
    "I seem to be higher on this film than many reviewers, but lower on it than I hoped I would be. I'd rate it a 6/10. I'll probably never watch it again, but I don't regret seeing it.The biggest problem isn't the acting, which is quite good, especially from Emily Blunt and Josh O'Conner. The problem is that the movie gradually loses focus on its most interesting elements. At its core, there are intriguing ideas about understanding, empathy, and communication, and those moments were far more compelling to me than the extended chase sequences and action set pieces.As the film progresses, it becomes increasingly focused on spectacle. The visual effects are impressive, and some of the action scenes are exciting, but they often come at the expense of character development and emotional investment. I found myself wishing the film had spent more time exploring the two main characters and why their journey mattered.That said, there is still plenty to admire. John Williams' score is outstanding and adds a sense of wonder that elevates many scenes. Spielberg's craftsmanship remains evident throughout, and several moments contain the sense of curiosity and humanity that have always made his best films memorable. Is it essential viewing? No. Is it the disaster some reviews make it out to be? Not even close. It's an uneven film with strong performances, a beautiful score, and an interesting message at its heart. If you enjoy science fiction and have the opportunity to see it on a large screen, it's worth experiencing, even if it ultimately falls short of its potential.",
]
proc = [preprocess(s) for s in samples]
seqs = tokenizer.texts_to_sequences(proc)
x = pad_sequences(seqs, padding='post', maxlen=MAXLEN)

# 4) predict
probs = model.predict(x)         # shape (n,1) with sigmoid output
preds = (probs.ravel() >= 0.5).astype(int)  # 1=positive, 0=negative

for s, p, pr in zip(samples, preds, probs.ravel()):
    label = "positive" if p==1 else "negative"
    print(f"Text: {s}\n  prob: {pr:.4f} → {label}\n")
```
